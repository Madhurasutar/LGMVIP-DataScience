```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
data=pd.read_csv("C:/Users/shree/Downloads/globalterrorismdb_0718dist.csv/Terrorism dataset.csv")
warnings.filterwarnings('ignore')
data
```

```python
#Top 10 countries affected by terrorism
print("Most country have a terrorism is:",data["country_txt"].value_counts().index[0])
print("Most region have terrorism is:",data["region_txt"].value_counts().index[0])
print("Most years in which terrorism occured was among:",data["iyear"].value_counts().index[:5].sort)
```

```python
#Terrorism Activities Each Year
plt.subplots(figsize=(15,6))
sns.countplot(data['iyear'],data=data,palette='flare')
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Year")
plt.title("Terrorism activities each year")
plt.show()
warnings.filterwarnings("ignore")
```

```python
data_India=data[data["country_txt"]=="India"]
```

```python
data_India.head()
```

```python
#Terrorism Activities Each Year In India
plt.subplots(figsize=(16,6))
sns.countplot(data_India["iyear"],data=data,palette="flare")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Year")
plt.title("Terrorism Activities Each Year")
plt.show()
warnings.filterwarnings("ignore")
```

```python
#Number of terrorist activities by region
plt.subplots(figsize=(15,6))
sns.countplot('region_txt',data=data,palette='RdYlGn',edgecolor=sns.color_palette('dark',4),order= data['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel("Regions")
plt.title('Number Of Terrorist Activities By Region')
plt.show()
warnings.filterwarnings("ignore")

              
```

```python
#Number of Terrorist Targets in India
plt.subplots(figsize=(15,6))
sns.countplot(data_India['targtype1_txt'], data=data, palette='RdBu',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.xlabel("Target Area")
plt.title('Number Of Terrorist Targets')
plt.show()
warnings.filterwarnings("ignore")
```

```python
#Succesful and failure operation in India
data_copy=data_India.copy()
data_copy["success"]=data_India["success"].replace({1:"successful operation",0:"failure operation"})
data_copy["success"].value_counts().plot(kind="pie",autopct="%1.1f%%",cmap="Accent")
plt.title("Successfuk & Failure Operation")
```

```python
# Most countries affected by terrorism
plt.subplots(figsize=(15,8))
sns.countplot('country_txt',data=data,palette='inferno',order=data['country_txt'].value_counts().index[:10])
plt.xticks(rotation=90)
plt.title("Most countries affected by terrorism")
plt.xlabel("Country")
plt.grid(axis="y")
plt.show()
warnings.filterwarnings('ignore')
```

```python
#The most active terror groups
sns.barplot(data['gname'].value_counts()[:5].values,data['gname'].value_counts()[:5].index)
plt.xticks(rotation=0)
fig=plt.gcf()
fig.set_size_inches(22,10)
plt.title("The Most Active Terror Groups")
plt.show()
warnings.filterwarnings("ignore")
```
