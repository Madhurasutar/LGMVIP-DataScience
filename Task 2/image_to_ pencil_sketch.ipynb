```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To covert image to RGB format
image=cv2.imread("C:/Users/shree/Downloads/image.webp")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Converting image to gray scale
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Inverting the image
inv_img=255-gray_img
plt.imshow(cv2.cvtColor(inv_img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
#smoothing or Blurr image
blurred=cv2.GaussianBlur(inv_img,(21,21),0)
```

```python
#Inverting the blurred image
inv_blur=255-blurred
pencil=cv2.divide(gray_img,inv_blur,scale=256.0)
plt.imshow(cv2.cvtColor(pencil,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
